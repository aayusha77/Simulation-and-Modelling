# Seed (in Pseudo-Random Number Generation)

A seed is an initial value used to start a pseudo-random number generator (PRNG). Since computers cannot generate truly random numbers, they use mathematical formulas to produce sequences that appear random.

The seed determines the entire sequence of generated numbers. If the same seed is used again, the same sequence will always be produced.

Thus, randomness in computers is deterministic but appears random.

# Seed Process
- Select initial seed value r0
- Use it in a recurrence formula
- Generate sequence step-by-step
- Output is deterministic for same seed

# Random State

Random state is the implementation of seed in machine learning libraries (like NumPy, Scikit-learn, Pandas). It ensures that operations involving randomness (like shuffling or splitting data) produce repeatable results.

It is essentially a controlled seed parameter.

# Random State Process
- Set random_state value
- Library initializes internal RNG using this value
- Random operations follow deterministic sequence
- Same random_state → same results every time

# Chi-Square Test

The Chi-Square (χ²) Test is a statistical test used to determine whether a set of observed frequencies differs significantly from the expected frequencies. In random number generation, the Chi-Square Test is used to evaluate the uniformity of pseudo-random numbers. If the generated numbers are uniformly distributed, the observed frequencies in different intervals should be close to the expected frequencies.

The generated random numbers are divided into several classes (intervals), and the frequency of numbers in each class is counted. The calculated Chi-Square value is then compared with a critical value from the Chi-Square distribution table. If the calculated value is less than the critical value, the random numbers are considered to be uniformly distributed.

# Formula

# Chi-Square Statistic:

χ² = Σ [(Oᵢ − Eᵢ)² / Eᵢ]

Notations

χ² = Chi-Square test statistic

Σ = Summation over all classes

Oᵢ = Observed frequency of the iᵗʰ class

Eᵢ = Expected frequency of the iᵗʰ class

i = Class index

Expected Frequency

Eᵢ = N / k

Where:

N = Total number of random numbers generated

k = Number of classes (intervals)

Degree of Freedom

df = k − 1

Where:

df = Degrees of freedom

k = Number of classes

Decision Rule

If χ²(calculated) < χ²(table), accept the null hypothesis (numbers are uniformly distributed).

If χ²(calculated) ≥ χ²(table), reject the null hypothesis (numbers are not uniformly distributed).

# 2. Random State Example (Scikit-learn)

```python
from sklearn.model_selection import train_test_split

# Sample dataset
X = [[1], [2], [3], [4], [5]]
y = [10, 20, 30, 40, 50]

# With random_state
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42
)

print("X_train:", X_train)
print("X_test:", X_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2
# Constants
iterator = 256 # Number of random numbers to generate
a = 29 # Multiplier
b = 37 # Increment
p = 256 # Modulus
n = 8 # Number of bins for Chi-Square test
x02 = 0 # Chi-Square statistic initialization
critical_value = chi2.pdf(1 - 0.05, df = n - 1) # Chi-Square critical value
# Initialize arrays
r = np.ones(shape=(iterator), dtype=int) # Array to store generated random numbers
ei = iterator / n # Expected frequency for each bin
c = np.zeros(shape=(n), dtype=int) # Bin counts
# Initial seed
r[0] = 17
# Generate random numbers using LCG
for i in range(0, iterator - 1):
 r[i + 1] = (r[i] * a + b) % p
# Bin the generated numbers and compute frequencies
for i in range(iterator):
    if r[i] <= 31:
        c[0] += 1
    elif r[i] <= 63:
       c[1] += 1
    elif r[i] <= 95:
        c[2] += 1
    elif r[i] <= 127:
        c[3] += 1
    elif r[i] <= 159:
        c[4] += 1
    elif r[i] <= 191:
        c[5] += 1
    elif r[i] <= 223:
        c[6] += 1
    elif r[i] <= 255:
        c[7] += 1
# Compute Chi-Square statistic
for i in range(n):
 x02 += ((c[i] - ei) ** 2) / ei
# Print results
print("Values of r:", r)
print("Values of c (bin frequencies):", c)
print("Value of Chi-Square statistic (x02):", x02)
print("The critical value is", critical_value)
print("Chi square test passed" if x02 <= critical_value else "Chi squsre test failed")
# Visualize the generated random numbers
plt.figure()
plt.plot(r, label="Generated Sequence (r)")
plt.xlabel("Index")
plt.ylabel("Random Number")
plt.title("LCG Generated Sequence")
plt.legend()
plt.show()
```

# Discussion

In this experiment, pseudo-random numbers were generated using the Linear Congruential Generator (LCG). The original implementation used parameters a=29, b=37, p=256, and seed X0=17. The generated sequence was divided into 8 bins, and the Chi-Square test was applied to check the uniformity of the distribution.

In the original case, since the modulus value was larger (p=256), the generated numbers covered a wider range (0–255). However, the distribution of values across bins was not perfectly uniform, leading to some variation in observed frequencies. This resulted in a comparatively higher Chi-Square statistic, indicating moderate randomness quality.

After modifying the parameters (changing a=17, b=43, p=128, and seed X0=21), the range of generated numbers became smaller and more controlled (0–127). This change improved the spread of values across the bins, reducing clustering in specific intervals. As a result, the observed frequencies became closer to expected frequencies, and the Chi-Square statistic decreased compared to the original case. This indicates an improvement in uniformity and randomness quality.

# Conclusion

The experiment demonstrates that the quality of pseudo-random numbers generated using the Linear Congruential Generator (LCG) is highly dependent on the choice of parameters. The original parameter set produced a less uniform distribution, while the modified parameters improved the distribution of values across bins.

The Chi-Square test confirmed this improvement by showing a lower test statistic in the modified case compared to the original. Therefore, the modified LCG configuration provides better uniformity and more reliable pseudo-random number generation.
